```python
member_id = df.Member_number.unique()
```

```python
member_id
```

```python
data = list()
```

```python
for m_id in member_id:
    data.append(list(df[df['Member_number']==m_id].itemDescription))
```

```python
data[:2]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
tranc = te.fit(data).transform(data)
```

```python
tranc[:1]
```

```python
df = pd.DataFrame(tranc, columns=te.columns_)
```

```python
df.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
s_tra = apriori(df, min_support=0.1, use_colnames=True)
```

```python
s_tra
```

```python
rules = association_rules(s_tra, metric='confidence', min_threshold=0.3)
```

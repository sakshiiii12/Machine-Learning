<h2><b>Dimension Reduction</b></h2>
<p>It's the process of reducing the features(inputs) of a dataset</p>

<h4><b>1. Linear Model / Methods</b></h4>
<table border>
    <tr>
        <th>Methods</th>
        <th>Description</th>
        <th>Ideal for use case</th>
    </tr>
    <tr>
        <td>PCA (Principle Component Analysis)</td>
        <td>Dataset has maximum variance</td>
        <td>Continous Values / Numeric Values</td>
    </tr>
    <tr>
        <td>LDA (Linear Discriminant Analysis)</td>
        <td>Super Version of PCA</td>
        <td>Classification / Labeled Dataset</td>
    </tr>
    <tr>
        <td>SVD (Singular Value Decomposition)</td>
        <td>Good for NLP</td>
        <td>Image/NLP Decomposition</td>
    </tr>
</table>

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
X = data.data
```

```python
y = data.target
```

```python
# PCA on Iris dataset
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(X)
```

```python
plt.scatter(x_pca[:,0] , x_pca[:,1] , c = y)
plt.show()
```

```python
# LDA on Iris dataset
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
x_lda = lda.fit_transform(X, y)
```

```python
plt.scatter(x_lda[:,0] , x_lda[:,1] , c = y)
plt.show()
```

```python
# t-SNE 
from sklearn.manifold import TSNE
```

```python
t_sne = TSNE(n_components=2)
```

```python
x_tsne = t_sne.fit_transform(X)
```

```python
plt.scatter(x_tsne[:,0] , x_tsne[:,1] , c = y)
plt.show()
```

```python
# UMap 
import umap.umap_ as umap
```

```python
um = umap.UMAP(n_components=2)
```

```python
x_umap = um.fit_transform(X)
```

```python
plt.scatter(x_umap[:,0] , x_umap[:,1] , c = y)
plt.show()
```

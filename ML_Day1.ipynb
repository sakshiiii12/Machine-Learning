```python
#Finding missing value's columns name and data type
missing_col = []
for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(col, "\tDataType:",df[col].dtype )
        missing_col.append(col)
```

```python
missing_col
```

```python
# Fiiling missing values (object)
for col in missing_col:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(value = df[col].mode()[0])
```

```python
# Fiiling missing values (float64)
for col in missing_col:
    if df[col].dtype == 'float64':
        df[col] = df[col].fillna(value = df[col].mean())
```

```python
def convertVotes(val):
    val = str(val)
    if val[-1] == 'K':
        return int(float(val[0:len(val)-1]))*1000
    elif val[-1] == 'M':
        return int(float(val[0:len(val)-1]))*1000000
    else:
        return int(val)
```

```python
df2.Votes = df2.Votes.map(convertVotes)
```

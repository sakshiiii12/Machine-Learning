# Random Forest Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
df = pd.read_csv("c:/users/sakshi yadav/Downloads/Iris_dataset.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.drop(columns=['Id'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.Species.unique()
```

```python
dic = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
def encode(val):
    return dic[val]
```

```python
df.Species = df.Species.map(encode)
```

```python
df.head()
```

```python
X = df.drop(columns=['Species'])
```

```python
y = df.Species
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy Score: ",round(accuracy_score(y_pred,y_test)*100),'%')
```
